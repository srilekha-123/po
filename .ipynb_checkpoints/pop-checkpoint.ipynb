```python
#finding missing values in the data
print("Finding missing values in cutomerdf:")
print(customerdf.isnull().sum()) # it checks if there are any missing columns in each column
print("Finding missing values in transactiondf:")
print(transactiondf.isnull().sum())
```

```python
#check unique values - customerdf
# print(customerdf.head())
# print(customerdf['customer_id'].unique()) - it gives all the unique values
# print(customerdf['customer_id'].value_counts()) - it gives count of each values in the columns
# customerdf['customer_id'].duplicated() - it gives false if the values is not duplicated
#nunique - gives no of unique values in the particular column
print(customerdf.count())
cols=customerdf.columns
print("List of columns present in customerdf:")
print(cols)
print("Number of unique values present in each column")
for i in cols:
    print(f"{i}     : ",customerdf[i].nunique())
```

```python
# check unique values - transactiondf
print(transactiondf.count())
print("List of columns present in transactiondf:")
cols=transactiondf.columns
print(cols)
print("Number of unique values present in each column")
for i in cols:
    print(f"{i} : ",transactiondf[i].nunique())
```

```python
#customerdf
import matplotlib.pyplot as plt
country_c=customerdf['country'].value_counts()
city_c=customerdf['city'].value_counts()
print(country_c)
```

```python
import seaborn as sns
plt.figure(figsize=(12,12))
plt.pie(country_c,labels=country_c.index, autopct='%1.1f%%',colors=sns.color_palette('viridis', len(country_c)))
plt.title("Customer distributiob by Country")
# plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=city_c.index,y=city_c.values,palette='viridis')
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of Customers by City')
plt.xticks(rotation=80)
plt.show()
```

```python
#transactiondf
#product category
prodcate_c=transactiondf['product_category'].value_counts()
print(prodcate_c)
```

```python
plt.figure(figsize=(10,6))
plt.bar(prodcate_c.index,prodcate_c)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(prodcate_c,autopct='%1.1f%%')
plt.legend(labels=prodcate_c.index,loc='upper right',bbox_to_anchor=(1.5,0.8))
plt.title("Products ditribution by Category")
plt.show()
```

```python
#product category
payment_c=transactiondf['payment_type'].value_counts()
print(payment_c)
```

```python
plt.figure(figsize=(6,6))
plt.pie(payment_c,autopct='%1.1f%%')
plt.legend(labels=payment_c.index,loc='upper right',bbox_to_anchor=(1.5,0.8))
plt.title("Payment Distribution")
plt.show()
```

```python
#ecommerce website
ecom_c=transactiondf['ecommerce_website_name'].value_counts()
print(ecom_c)
```
